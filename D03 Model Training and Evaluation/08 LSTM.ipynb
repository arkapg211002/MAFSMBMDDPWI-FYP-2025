# LSTM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import LabelEncoder, label_binarize
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import StratifiedKFold
from itertools import cycle

# Load the dataset
data = pd.read_csv('preprocessed_mental_health.csv')

# Separate features and target
X = data['text']
y = data['mental_health_issue']

# Encode target labels
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y = to_categorical(y)  # Convert labels to one-hot encoded format for multi-class classification

# Tokenize and pad the text sequences
vocab_size = 25000  # Set a vocabulary size
max_length = 128   # Set a max length for padding

tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(X)

X_seq = tokenizer.texts_to_sequences(X)
X_padded = pad_sequences(X_seq, maxlen=max_length, padding='post', truncating='post')

# Define Stratified K-Fold cross-validation
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize variables to store the overall accuracy, confusion matrix, and ROC-AUC data
all_accuracy = []
all_cm = []
val_losses = []  # To store validation loss for each fold
val_accuracies = []  # To store validation accuracy for each fold
all_y_true = []  # To store true labels
all_y_pred = []  # To store predicted probabilities

# Cross-validation loop
for train_index, val_index in kf.split(X_padded, np.argmax(y, axis=1)):
    X_train, X_val = X_padded[train_index], X_padded[val_index]
    y_train, y_val = y[train_index], y[val_index]

    # Build the LSTM model
    model = Sequential([
        Embedding(vocab_size, 128, input_length=max_length),
        LSTM(128, return_sequences=True),
        Dropout(0.2),
        LSTM(64),
        Dropout(0.2),
        Dense(64, activation='relu'),
        Dense(y.shape[1], activation='softmax')  # Output layer with softmax for multi-class classification
    ])

    model.compile(optimizer='adamw', loss='categorical_crossentropy', metrics=['accuracy'])

    # Train the model
    history = model.fit(X_train, y_train, epochs=20, batch_size=16, validation_data=(X_val, y_val), verbose=1)

    # Evaluate the model on the validation set
    val_loss, val_accuracy = model.evaluate(X_val, y_val, verbose=0)
    all_accuracy.append(val_accuracy)
    val_losses.append(val_loss)
    val_accuracies.append(val_accuracy)

    # Get predictions (probabilities)
    y_pred = model.predict(X_val)

    # Collect true labels and predicted probabilities
    all_y_true.append(y_val)
    all_y_pred.append(y_pred)

    # Generate predictions and convert back from one-hot encoding
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_val_classes = np.argmax(y_val, axis=1)

    # Calculate confusion matrix for this fold
    cm = confusion_matrix(y_val_classes, y_pred_classes)
    all_cm.append(cm)

# Flatten the list of true labels and predictions
all_y_true = np.concatenate(all_y_true, axis=0)
all_y_pred = np.concatenate(all_y_pred, axis=0)

# Binarize the true labels for ROC-AUC calculation
y_true_binarized = label_binarize(np.argmax(all_y_true, axis=1), classes=range(y.shape[1]))
y_pred_binarized = all_y_pred  # No need to binarize predictions as they are already probabilities

# Initialize ROC data
fpr = {}
tpr = {}
roc_auc = {}

# Calculate ROC-AUC for each class
for i in range(y.shape[1]):
    fpr[i], tpr[i], _ = roc_curve(y_true_binarized[:, i], y_pred_binarized[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_true_binarized.ravel(), y_pred_binarized.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Plot ROC-AUC curves
plt.figure(figsize=(10, 8))
colors = cycle(['blue', 'red', 'green', 'orange', 'purple'])

for i, color in zip(range(y.shape[1]), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'Class {label_encoder.classes_[i]} (AUC = {roc_auc[i]:.2f})')

# Plot micro-average ROC curve
plt.plot(fpr["micro"], tpr["micro"], color='deeppink', linestyle=':', linewidth=4,
         label=f'Micro-average (AUC = {roc_auc["micro"]:.2f})')

# Plot the diagonal line for random guessing
plt.plot([0, 1], [0, 1], 'k--', lw=2)

plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.title('Multi-Class ROC-AUC Curve', fontsize=16)
plt.legend(loc='lower right', fontsize=12)
plt.grid(alpha=0.3)
plt.show()

# Plot val_loss and val_accuracy across folds
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(val_losses) + 1), val_losses, label='Validation Loss', marker='o')
plt.plot(range(1, len(val_accuracies) + 1), val_accuracies, label='Validation Accuracy', marker='o')
plt.title('Validation Loss and Accuracy Across Folds')
plt.xlabel('Fold')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.show()

# Calculate average accuracy across all folds
avg_accuracy = np.mean(all_accuracy)
print(f'Average Cross-validated Accuracy: {avg_accuracy * 100:.2f}%')

# Compute the average confusion matrix
avg_cm = np.mean(all_cm, axis=0).astype(int)

# Print classification report for the last fold (optional)
print("Classification Report:\n", classification_report(np.argmax(all_y_true, axis=1), np.argmax(all_y_pred, axis=1), target_names=label_encoder.classes_))

# Plot the average confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(avg_cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Average Confusion Matrix Across All Folds')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming `model` is your trained LSTM model and `X_padded` is the tokenized input

# Get the embedding layer weights
embedding_layer = model.get_layer(index=0)  # Embedding layer is the first layer
embedding_weights = embedding_layer.get_weights()[0]  # Shape: (vocab_size, embedding_dim)

# Generate random embeddings to simulate untrained embeddings
random_embeddings = np.random.normal(size=embedding_weights.shape)  # Random initialization

# Select the number of words to plot for better visualization
num_words_to_plot = 500

# Create a wide figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(16, 8))  # Wide aspect ratio

# Plot Random Embeddings
sns.heatmap(
    random_embeddings[:num_words_to_plot],
    cmap='viridis',
    ax=axes[0],
    cbar=True
)
axes[0].set_title('Random Embeddings Heatmap (Before Training)', fontsize=14)
axes[0].set_xlabel('Embedding Dimensions', fontsize=12)
axes[0].set_ylabel('Words', fontsize=12)

# Plot Learned Embeddings
sns.heatmap(
    embedding_weights[:num_words_to_plot],
    cmap='viridis',
    ax=axes[1],
    cbar=True
)
axes[1].set_title('Learned Embeddings Heatmap (After Training)', fontsize=14)
axes[1].set_xlabel('Embedding Dimensions', fontsize=12)
axes[1].set_ylabel('Words', fontsize=12)

# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```
